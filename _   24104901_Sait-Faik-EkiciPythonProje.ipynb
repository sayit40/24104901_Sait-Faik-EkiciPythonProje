```python
# Rehber (Kişi Listesi) - Kişi sözlüklerini tutacak liste
rehber = []
```

```python
def yeni_id_olustur(liste):
    """
    Verilen listeyi kontrol ederek yeni ve benzersiz bir ID oluşturur.
    Listede eleman yoksa 1 döner, varsa mevcut en yüksek ID'nin 1 fazlasını döner.
    """
    if len(liste) == 0:
        return 1
    else:
        en_buyuk_id = 0
        for eleman in liste:
            if eleman["id"] > en_buyuk_id:
                en_buyuk_id = eleman["id"]
        return en_buyuk_id + 1

def telefon_format_kontrol(telefon):
    """
    Telefon numarasının tam 10 haneli ve sadece rakamlardan
    oluşup oluşmadığını kontrol eder.
    """
    if len(telefon) == 10 and telefon.isdigit():
        return True
    return False

def email_format_kontrol(email):
    """
    E-posta adresinin içinde '@' ve '.' karakterlerinin olup olmadığını kontrol eder.
    """
    if "@" in email and "." in email:
        return True
    return False

def menu_goster():
    """
    Kullanıcıya programın ana menüsünü ekrana yazdırır.
    """
    print("\n" + "="*35)
    print("KİŞİSEL REHBER")
    print("="*35)
    print("REHBER İŞLEMLERİ")
    print("1. Yeni Kişi Ekle")
    print("2. Tüm Kişileri Listele")
    print("3. Kişi Ara")
    print("4. Kişi Sil")
    print("-" * 35)
```

```python
def kisi_ekle():
    """
    Kullanıcıdan bilgileri alır, format kontrollerini yapar ve doğruysa rehbere ekler.
    """
    print("\n--- Yeni Kişi Ekle ---")
    ad = input("Adınız: ")
    soyad = input("Soyadınız: ")
    
    # Doğru telefon girilene kadar sormaya devam et
    while True:
        telefon = input("Telefon (10 haneli, sadece rakam): ")
        if telefon_format_kontrol(telefon):
            break
        else:
            print("Hatalı giriş! Telefon 10 haneli olmalı ve harf içermemelidir.")
            
    # Doğru e-posta girilene kadar sormaya devam et
    while True:
        email = input("E-posta adresi: ")
        if email_format_kontrol(email):
            break
        else:
            print("Hatalı giriş! Lütfen geçerli bir e-posta adresi giriniz (@ ve . içermeli).")
            
    yeni_kisi = {
        "id": yeni_id_olustur(rehber),
        "ad": ad,
        "soyad": soyad,
        "telefon": telefon,
        "email": email
    }
    
    rehber.append(yeni_kisi)
    print(f"Başarılı: {ad} {soyad} rehbere eklendi.")

def kisileri_listele():
    """
    Rehberdeki tüm kişileri düzenli bir tablo formatında ekrana basar.
    """
    print("\n--- Tüm Kişiler ---")
    if len(rehber) == 0:
        print("Rehberinizde kayıtlı kimse bulunmamaktadır.")
        return
        
    print(f"{'ID':<4} | {'Ad':<12} | {'Soyad':<12} | {'Telefon':<12} | {'E-Posta'}")
    print("-" * 70)
    for kisi in rehber:
        print(f"{kisi['id']:<4} | {kisi['ad']:<12} | {kisi['soyad']:<12} | {kisi['telefon']:<12} | {kisi['email']}")

def kisi_ara():
    """
    Kullanıcıdan alınan kelimenin ad, soyad veya telefon içinde 
    geçip geçmediğine göre eşleşen kişileri listeler.
    """
    aranan = input("\nAramak istediğiniz kelime (Ad, Soyad veya Telefon): ").lower()
    bulunan_sayisi = 0
    
    print(f"\n--- Arama Sonuçları ---")
    print(f"{'ID':<4} | {'Ad':<12} | {'Soyad':<12} | {'Telefon':<12} | {'E-Posta'}")
    print("-" * 70)
    
    for kisi in rehber:
        # Arama yaparken büyük/küçük harf duyarlılığını kaldırmak için .lower() kullanıyoruz
        if aranan in kisi["ad"].lower() or aranan in kisi["soyad"].lower() or aranan in kisi["telefon"]:
            print(f"{kisi['id']:<4} | {kisi['ad']:<12} | {kisi['soyad']:<12} | {kisi['telefon']:<12} | {kisi['email']}")
            bulunan_sayisi += 1
            
    if bulunan_sayisi == 0:
        print("Arama kriterinize uygun kimse bulunamadı.")

def kisi_sil():
    """
    Kullanıcıdan bir ID alır ve bu ID'ye sahip kişiyi listeden çıkartır.
    """
    kisileri_listele()
    if len(rehber) == 0:
        return
        
    try:
        silinecek_id = int(input("\nSilmek istediğiniz kişinin ID numarasını giriniz: "))
        silindi_mi = False
        
        # Liste içinde dönerek ID eşleşmesini kontrol ediyoruz
        for i in range(len(rehber)):
            if rehber[i]["id"] == silinecek_id:
                silinen = rehber.pop(i)
                print(f"Başarılı: {silinen['ad']} {silinen['soyad']} rehberden silindi.")
                silindi_mi = True
                break
                
        if not silindi_mi:
            print("Hata: Bu ID numarasına sahip bir kişi bulunamadı.")
    except ValueError:
        print("Hata: Lütfen sadece rakamdan oluşan bir ID giriniz!")
```

```python
def main():
    """
    Programın ana döngüsünü oluşturur. Menüyü gösterir ve kullanıcının
    seçimine göre uygun fonksiyonu tetikler.
    """
    while True:
        menu_goster()
        secim = input("Seçiminiz: ")
        
        if secim == "1":
            kisi_ekle()
        elif secim == "2":
            kisileri_listele()
        elif secim == "3":
            kisi_ara()
        elif secim == "4":
            kisi_sil()
        elif secim == "0":
            print("Program sonlandırılıyor. Hoşçakalın!")
            break
        else:
            print("Geçersiz seçim! Lütfen 0 ile 4 arasında bir numara giriniz.")  
```
