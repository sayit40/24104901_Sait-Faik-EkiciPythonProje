## 2. Kullanılan Veri Yapıları
Bu bölümde projede kullanılacak global veri yapıları olan `rehber` ve `alisveris_listesi` tanımlanmıştır. Her iki liste de içerisinde sözlük (dictionary) veri tipinde elemanlar barındıracaktır.

```python
# Rehber (Kişi Listesi) - Kişi sözlüklerini tutacak liste
rehber = []

# Alışveriş Listesi - Ürün sözlüklerini tutacak liste
alisveris_listesi = []
```

## 3. Yardımcı Fonksiyonlar
ID oluşturma, telefon ve e-posta formatı doğrulama ile ana menüyü gösterme işlemleri için kullanılacak yardımcı fonksiyonlar bu bölümde yer almaktadır.

```python
def yeni_id_olustur(liste):
    """
    Verilen listeyi kontrol ederek yeni ve benzersiz bir ID oluşturur.
    Listede eleman yoksa 1 döner, varsa mevcut en yüksek ID'nin 1 fazlasını döner.
    """
    if len(liste) == 0:
        return 1
    else:
        en_buyuk_id = 0
        for eleman in liste:
            if eleman["id"] > en_buyuk_id:
                en_buyuk_id = eleman["id"]
        return en_buyuk_id + 1

def telefon_format_kontrol(telefon):
    """
    Telefon numarasının tam 10 haneli ve sadece rakamlardan
    oluşup oluşmadığını kontrol eder.
    """
    if len(telefon) == 10 and telefon.isdigit():
        return True
    return False

def email_format_kontrol(email):
    """
    E-posta adresinin içinde '@' ve '.' karakterlerinin olup olmadığını kontrol eder.
    """
    if "@" in email and "." in email:
        return True
    return False

def menu_goster():
    """
    Kullanıcıya programın ana menüsünü ekrana yazdırır.
    """
    print("\n" + "="*35)
    print("KİŞİSEL REHBER & ALIŞVERİŞ LİSTESİ")
    print("="*35)
    print("REHBER İŞLEMLERİ")
    print("1. Yeni Kişi Ekle")
    print("2. Tüm Kişileri Listele")
    print("3. Kişi Ara")
    print("4. Kişi Sil")
    print("-" * 35)
    print("ALIŞVERİŞ LİSTESİ İŞLEMLERİ")
    print("5. Yeni Ürün Ekle")
    print("6. Tüm Ürünleri Listele")
    print("7. Ürünü Alındı Olarak İşaretle")
    print("8. Alınmamış Ürünleri Listele")
    print("-" * 35)
    print("0. Çıkış")
    print("=" * 35)
```

## 4. Rehber Yönetim Fonksiyonları
Rehbere yeni kişi ekleme, mevcut kişileri tablo halinde listeleme, kişi arama ve kişi silme işlemlerini gerçekleştiren fonksiyonlar bu hücrede tanımlanmıştır.

```python
def kisi_ekle():
    """
    Kullanıcıdan bilgileri alır, format kontrollerini yapar ve doğruysa rehbere ekler.
    """
    print("\n--- Yeni Kişi Ekle ---")
    ad = input("Adınız: ")
    soyad = input("Soyadınız: ")
    
    # Doğru telefon girilene kadar sormaya devam et
    while True:
        telefon = input("Telefon (10 haneli, sadece rakam): ")
        if telefon_format_kontrol(telefon):
            break
        else:
            print("Hatalı giriş! Telefon 10 haneli olmalı ve harf içermemelidir.")
            
    # Doğru e-posta girilene kadar sormaya devam et
    while True:
        email = input("E-posta adresi: ")
        if email_format_kontrol(email):
            break
        else:
            print("Hatalı giriş! Lütfen geçerli bir e-posta adresi giriniz (@ ve . içermeli).")
            
    yeni_kisi = {
        "id": yeni_id_olustur(rehber),
        "ad": ad,
        "soyad": soyad,
        "telefon": telefon,
        "email": email
    }
    
    rehber.append(yeni_kisi)
    print(f"Başarılı: {ad} {soyad} rehbere eklendi.")

def kisileri_listele():
    """
    Rehberdeki tüm kişileri düzenli bir tablo formatında ekrana basar.
    """
    print("\n--- Tüm Kişiler ---")
    if len(rehber) == 0:
        print("Rehberinizde kayıtlı kimse bulunmamaktadır.")
        return
        
    print(f"{'ID':<4} | {'Ad':<12} | {'Soyad':<12} | {'Telefon':<12} | {'E-Posta'}")
    print("-" * 70)
    for kisi in rehber:
        print(f"{kisi['id']:<4} | {kisi['ad']:<12} | {kisi['soyad']:<12} | {kisi['telefon']:<12} | {kisi['email']}")

def kisi_ara():
    """
    Kullanıcıdan alınan kelimenin ad, soyad veya telefon içinde 
    geçip geçmediğine göre eşleşen kişileri listeler.
    """
    aranan = input("\nAramak istediğiniz kelime (Ad, Soyad veya Telefon): ").lower()
    bulunan_sayisi = 0
    
    print(f"\n--- Arama Sonuçları ---")
    print(f"{'ID':<4} | {'Ad':<12} | {'Soyad':<12} | {'Telefon':<12} | {'E-Posta'}")
    print("-" * 70)
    
    for kisi in rehber:
        # Arama yaparken büyük/küçük harf duyarlılığını kaldırmak için .lower() kullanıyoruz
        if aranan in kisi["ad"].lower() or aranan in kisi["soyad"].lower() or aranan in kisi["telefon"]:
            print(f"{kisi['id']:<4} | {kisi['ad']:<12} | {kisi['soyad']:<12} | {kisi['telefon']:<12} | {kisi['email']}")
            bulunan_sayisi += 1
            
    if bulunan_sayisi == 0:
        print("Arama kriterinize uygun kimse bulunamadı.")

def kisi_sil():
    """
    Kullanıcıdan bir ID alır ve bu ID'ye sahip kişiyi listeden çıkartır.
    """
    kisileri_listele()
    if len(rehber) == 0:
        return
        
    try:
        silinecek_id = int(input("\nSilmek istediğiniz kişinin ID numarasını giriniz: "))
        silindi_mi = False
        
        # Liste içinde dönerek ID eşleşmesini kontrol ediyoruz
        for i in range(len(rehber)):
            if rehber[i]["id"] == silinecek_id:
                silinen = rehber.pop(i)
                print(f"Başarılı: {silinen['ad']} {silinen['soyad']} rehberden silindi.")
                silindi_mi = True
                break
                
        if not silindi_mi:
            print("Hata: Bu ID numarasına sahip bir kişi bulunamadı.")
    except ValueError:
        print("Hata: Lütfen sadece rakamdan oluşan bir ID giriniz!")
```

 5. Alışveriş Listesi Fonksiyonları
Alışveriş sepetine ürün ekleme, tüm ürünleri listeleme, alınmamış ürünleri gösterme, ürün silme ve ürünü alındı olarak işaretleme fonksiyonları burada tanımlanmıştır.

```python
def urun_ekle():
    """
    Kullanıcıdan ürün bilgilerini alır ve alisveris_listesi'ne ekler.
    Miktar için sadece sayısal değer kabul edilir.
    """
    print("\n--- Yeni Ürün Ekle ---")
    urun_adi = input("Ürün Adı: ")
    
    try:
        miktar = float(input("Miktar: "))
    except ValueError:
        print("Hata: Lütfen miktar için geçerli bir sayı giriniz!")
        return
        
    birim = input("Birim (Örn: Adet, Litre, Kg): ")
    
    yeni_urun = {
        "id": yeni_id_olustur(alisveris_listesi),
        "urun_adi": urun_adi,
        "miktar": miktar,
        "birim": birim,
        "alinma_durumu": False # Varsayılan olarak alınmadı kabul edilir
    }
    
    alisveris_listesi.append(yeni_urun)
    print(f"Başarılı: {urun_adi} alışveriş listesine eklendi.")

def urunleri_listele():
    """
    Alışveriş listesindeki tüm ürünleri ekrana yazdırır.
    Alınan ürünlerin yanına onay işareti koyar.
    """
    print("\n--- Alışveriş Listesi ---")
    if len(alisveris_listesi) == 0:
        print("Alışveriş listeniz şu an boş.")
        return
        
    print(f"{'ID':<4} | {'Durum':<8} | {'Ürün Adı':<15} | {'Miktar':<8} | {'Birim'}")
    print("-" * 55)
    
    for urun in alisveris_listesi:
        # Alınma durumuna göre ekranda gösterilecek işareti belirliyoruz
        durum_isareti = "[ ✓ ]" if urun["alinma_durumu"] else "[ X ]"
        print(f"{urun['id']:<4} | {durum_isareti:<8} | {urun['urun_adi']:<15} | {urun['miktar']:<8} | {urun['birim']}")

def urun_tamamla():
    """
    Kullanıcının girdiği ID'deki ürünün durumunu günceller.
    """
    alinmamis_urunleri_listele()
    if len(alisveris_listesi) == 0:
        return
        
    try:
        tamam_id = int(input("\nAlındı olarak işaretlemek istediğiniz ürünün ID numarasını giriniz: "))
        bulundu = False
        
        for urun in alisveris_listesi:
            if urun["id"] == tamam_id:
                urun["alinma_durumu"] = True
                print(f"Başarılı: '{urun['urun_adi']}' alındı olarak işaretlendi.")
                bulundu = True
                break
                
        if not bulundu:
            print("Hata: Girdiğiniz ID ile eşleşen bir ürün bulunamadı.")
    except ValueError:
        print("Hata: Lütfen ID numarası olarak geçerli bir sayı giriniz.")

def urun_sil():
    """
    Kullanıcıdan bir ID alır ve ilgili ürünü alışveriş listesinden tamamen kaldırır.
    """
    urunleri_listele()
    if len(alisveris_listesi) == 0:
        return
        
    try:
        sil_id = int(input("\nSilmek istediğiniz ürünün ID numarasını giriniz: "))
        silindi_mi = False
        
        for i in range(len(alisveris_listesi)):
            if alisveris_listesi[i]["id"] == sil_id:
                silinen = alisveris_listesi.pop(i)
                print(f"Başarılı: '{silinen['urun_adi']}' alışveriş listesinden silindi.")
                silindi_mi = True
                break
                
        if not silindi_mi:
            print("Hata: Bu ID numarasına ait bir ürün bulunamadı.")
    except ValueError:
        print("Hata: Lütfen geçerli bir tam sayı giriniz.")

def alinmamis_urunleri_listele():
    """
    Sadece alinma_durumu False olan eksik ürünleri listeler.
    """
    print("\n--- Alınmamış Ürünler ---")
    alinmamis_sayisi = 0
    
    print(f"{'ID':<4} | {'Ürün Adı':<15} | {'Miktar':<8} | {'Birim'}")
    print("-" * 45)
    
    for urun in alisveris_listesi:
        if not urun["alinma_durumu"]:
            print(f"{urun['id']:<4} | {urun['urun_adi']:<15} | {urun['miktar']:<8} | {urun['birim']}")
            alinmamis_sayisi += 1
            
    if alinmamis_sayisi == 0:
        if len(alisveris_listesi) > 0:
            print("Tebrikler! Listede alınmamış ürün kalmadı.")
        else:
            print("Liste boş olduğu için gösterilecek ürün yok.")
```

## 6. Ana Menü ve Program Döngüsü
Uygulamanın sürekli çalışmasını sağlayan ana fonksiyon `main()` burada yer almaktadır. Kullanıcı "0" seçeneğini seçene kadar döngü devam eder.

```python
def main():
    """
    Programın ana döngüsünü oluşturur. Menüyü gösterir ve kullanıcının
    seçimine göre uygun fonksiyonu tetikler.
    """
    while True:
        menu_goster()
        secim = input("Seçiminiz: ")
        
        if secim == "1":
            kisi_ekle()
        elif secim == "2":
            kisileri_listele()
        elif secim == "3":
            kisi_ara()
        elif secim == "4":
            kisi_sil()
        elif secim == "5":
            urun_ekle()
        elif secim == "6":
            urunleri_listele()
        elif secim == "7":
            urun_tamamla()
        elif secim == "8":
            alinmamis_urunleri_listele()
        elif secim == "0":
            print("Program sonlandırılıyor. Hoşçakalın!")
            break
        else:
            print("Geçersiz seçim! Lütfen 0 ile 8 arasında bir numara giriniz.")
```

## 7. Programın Çalıştırılması
Aşağıdaki hücreyi çalıştırdığınızda programın arayüzü başlatılacaktır.

```python
# Eğer bu kod dosyası doğrudan çalıştırılıyorsa main fonksiyonunu başlat
if __name__ == "__main__":
    try:
        main()
    except NameError:
        # main() hücresi çalıştırılmadan bu hücre çalıştırılırsa verilecek uyarı:
        print("Sistem Uyarısı: 'main' fonksiyonu henüz hafızada yok. Lütfen yukarıdaki 'def main():' hücresini çalıştırdığınızdan emin olun!")
    except KeyboardInterrupt:
        # Kullanıcı program çalışırken "Stop" butonuna basarsa verilecek uyarı:
        print("\nSistem Uyarısı: Program kullanıcı tarafından manuel olarak durduruldu. Çıkış yapılıyor...")
```
